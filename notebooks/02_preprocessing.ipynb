# 02 — Preprocessing
Scale features, split data, apply SMOTE, and save processed splits.

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
from src.preprocess import load_data, split_and_scale, apply_smote

df = load_data('../data/raw/creditcard.csv')
print('Loaded:', df.shape)
```

```python
X_train, X_test, y_train, y_test = split_and_scale(df)
print(f'Train: {X_train.shape}, Test: {X_test.shape}')
print(f'Train fraud rate: {y_train.mean():.4f}')
print(f'Test  fraud rate: {y_test.mean():.4f}')
```

```python
X_train_sm, y_train_sm = apply_smote(X_train, y_train)
print(f'After SMOTE — Train: {X_train_sm.shape}')
print(f'Class counts: {pd.Series(y_train_sm).value_counts().to_dict()}')
```

```python
# Save splits for use in model notebooks
import os
os.makedirs('../data/processed', exist_ok=True)

pd.DataFrame(X_train).to_csv('../data/processed/X_train.csv', index=False)
pd.DataFrame(X_test).to_csv('../data/processed/X_test.csv', index=False)
pd.Series(y_train).to_csv('../data/processed/y_train.csv', index=False)
pd.Series(y_test).to_csv('../data/processed/y_test.csv', index=False)

pd.DataFrame(X_train_sm).to_csv('../data/processed/X_train_smote.csv', index=False)
pd.Series(y_train_sm).to_csv('../data/processed/y_train_smote.csv', index=False)

print('Saved processed splits to data/processed/')
```
