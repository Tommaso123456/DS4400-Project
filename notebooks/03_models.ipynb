# 03 — Model Training
Train all four models on SMOTE-resampled data. Each model is saved immediately after training.

```python
import sys
sys.path.append('..')

import os
import pandas as pd
import pickle
from src.models import (
    train_logistic_regression,
    train_random_forest,
    train_xgboost,
    train_neural_network,
)

os.makedirs('../results', exist_ok=True)

X_train = pd.read_csv('../data/processed/X_train_smote.csv')
y_train = pd.read_csv('../data/processed/y_train_smote.csv').squeeze()
print(f'Training set: {X_train.shape}, fraud rate: {y_train.mean():.4f}')
```

```python
print('Training Logistic Regression...')
lr = train_logistic_regression(X_train, y_train)
with open('../results/lr.pkl', 'wb') as f:
    pickle.dump(lr, f)
print('Done — saved lr.pkl')
```

```python
print('Training Random Forest...')
rf = train_random_forest(X_train, y_train)
with open('../results/rf.pkl', 'wb') as f:
    pickle.dump(rf, f)
print('Done — saved rf.pkl')
```

```python
# XGBoost handles class imbalance via scale_pos_weight — use original (non-SMOTE) training data
# to avoid OOM on the ~454k oversampled dataset
X_train_orig = pd.read_csv('../data/processed/X_train.csv')
y_train_orig = pd.read_csv('../data/processed/y_train.csv').squeeze()
print(f'Original train set: {X_train_orig.shape}, fraud rate: {y_train_orig.mean():.4f}')

print('Training XGBoost...')
xgb = train_xgboost(X_train_orig, y_train_orig)
with open('../results/xgb.pkl', 'wb') as f:
    pickle.dump(xgb, f)
print('Done — saved xgb.pkl')
```

```python
print('Training Neural Network...')
nn = train_neural_network(X_train, y_train, epochs=20)
with open('../results/nn.pkl', 'wb') as f:
    pickle.dump(nn, f)
print('Done — saved nn.pkl')
```

```python
print('All models saved to results/')
print([f for f in os.listdir('../results') if f.endswith('.pkl')])
```
